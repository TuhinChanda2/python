```python
import numpy as np
import matplotlib.pyplot as plt
import random
class PythonGraphs:
    def Graph1():
        x_axis=np.random.randint(10,50,5)
        y_axis=np.random.randint(10,50,5)
        color_section=np.array(["red","blue","green","yellow","purple"])
        plt.bar(x_axis,y_axis,color=color_section,width=3)
        plt.plot(x_axis,y_axis,marker="o",mfc="red",mec="red")
        print("------Coordinates Values-----")
        print("x=>",x_axis)
        print("y=>",y_axis)
        plt.grid()
        # plt.legend(x_axis)
        # print(x_axis[0])     #to access element
        plt.show()
PythonGraphs.Graph1()
```
