```python
import numpy as np
import random
import matplotlib.pyplot as plt
class SubPlot:
    def GraphPlots():
        x_axis=np.random.randint(10,50,8)
        y_axis=np.random.randint(10,50,8)
        plt.subplot(2,2,1)
        plt.grid(color="green")
        plt.plot(x_axis,marker="o",mfc="red",color="blue")
        plt.plot(y_axis,marker="o",mfc="purple",color="black")
        

        a_axis=np.random.randint(10,40,8)
        b_axis=np.random.randint(10,30,8)
        plt.subplot(2,2,2)
        plt.grid()
        plt.barh(a_axis,b_axis)
        

        c_axis=np.random.randint(50,100,12)
        d_axis=np.random.randint(50,100,12)
        plt.subplot(2,2,3)
        plt.grid()
        plt.scatter(c_axis,d_axis,color="red")
        
        
        e_axis=np.random.randint(100,200,8)
        labelValue=['A','B','C','D','E','F','G','H']
        plt.subplot(2,2,4)
        plt.pie(e_axis,labels=labelValue,shadow=True)
        plt.legend()
        plt.show()
SubPlot.GraphPlots()
        
```
